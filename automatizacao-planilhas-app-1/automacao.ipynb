#### Planilha (xlsx): Folha de pagamento

```python
df = pd.read_excel('input\FolhaPagto-31082023_0011.xlsx', header=3)
df.head()
```

```python
colunas_para_alterar = df.columns[6:63]

# Função para tratar valores "-" como nulos (NaN)
def tratar_valor(valor):
    valor = str(valor).strip()
    if valor == '-':
        return 0
    return valor

# Loop pelas colunas e aplique a função de tratamento em cada uma delas.
for coluna in colunas_para_alterar:
    df[coluna] = df[coluna].apply(tratar_valor)

for coluna in colunas_para_alterar:
   df[coluna] = df[coluna].astype('float')

df.head(20)
```

```python
df.info()
```

```python
colunas_para_somar = df.iloc[:, 6:]
soma_colunas = colunas_para_somar.sum().round(2)
soma_total = soma_colunas.sum()

# Crie um novo DataFrame para mostrar a soma das colunas.
df_soma = pd.DataFrame({'Coluna': soma_colunas.index, 'Soma': soma_colunas.values})

print(soma_total)

print(df_soma)
```

##### Parte 3: Planilha resumo por CENTRO DE CUSTO

```python
df_codigos = pd.read_csv('utils\\codigos.csv', encoding='ISO-8859-1')
df_codigos.info()
```

```python
df_codigos['CENTRO DE CUSTO'].value_counts()
```

```python
def associar_valor(cr_desejado):
    data = resumo.set_index('CR')
    resultados = pd.DataFrame(columns=['DATA', 'TIPO', 'DEBITO', 'CREDITO', 'DESCRICAO COMPLETA', 'VALOR', 'UNIDADE DE NEGOCIO', 'CR'])

    if cr_desejado == 'CR1002':
        codigos = df_codigos.loc[df_codigos['CENTRO DE CUSTO'] == 'adm']
    elif cr_desejado == 'CR2020':
        codigos = df_codigos.loc[df_codigos['CENTRO DE CUSTO'] == 'com']
    else:
        codigos = df_codigos.loc[df_codigos['CENTRO DE CUSTO'] == 'custo']

    
    for index, linha in codigos.iterrows():
        codigo1_correspondente = linha['DEBITO']
        codigo2_correspondente = linha['CREDITO']
        descricao_correspondente = linha['DESCRICAO']
        folha_correspondente = linha['COLUNA DA FOLHA']
        
        if folha_correspondente in lista_resumo['LISTA'].values:
            valor_correspondente = data.loc[cr_desejado, folha_correspondente]
            unidade_correspondente = data.loc[cr_desejado, 'UNIDADE DE NEGOCIO']
            ref_completa = f'{descricao_correspondente} {folha_correspondente}'

            if valor_correspondente > 0:
                resultados.loc[len(resultados)] = ['31/07/2023', 'MANUAL', codigo1_correspondente, codigo2_correspondente, ref_completa, valor_correspondente, unidade_correspondente, cr_desejado]
            else:
                resultados = resultados
        else:
            resultados

    return resultados
```

```python
associar_valor('CR1002')
```

```python
resultados = (associar_valor(valor) for valor in resumo['CR'])
resultado_final = pd.concat(resultados, ignore_index=True)
```
